```python
import fitz  # PyMuPDF
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Function to read PDF and return its text
def get_text_from_pdf(Test):
    doc = fitz.open(Test)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("sshleifer/distilbart-cnn-12-6")
model = AutoModelForSeq2SeqLM.from_pretrained("sshleifer/distilbart-cnn-12-6")

def summarize_text(text, chunk_size=512, max_length=150, min_length=40):  # Adjusted chunk_size
    # Break the text into smaller chunks
    # Note: Now chunking raw text before tokenization to better control size
    raw_chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
    summaries = []
    for chunk in raw_chunks:
        # Ensure each chunk is within the token limit
        inputs = tokenizer(chunk, return_tensors="pt", truncation=True, max_length=1024, padding="max_length")
        summary_ids = model.generate(inputs.input_ids, max_length=max_length, min_length=min_length, length_penalty=2.0, num_beams=4, early_stopping=True)
        chunk_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        summaries.append(chunk_summary)

    return " ".join(summaries)

# Example PDF reading and summarization
pdf_path = "Test.pdf"  # Update this path
document_text = get_text_from_pdf(pdf_path)
summary = summarize_text(document_text)
print(summary)
```
